# Question 6
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. 

Label both lines and show the legend.
Don't forget to label your axes!

```python
import matplotlib.pyplot as plt
marstat = pd.read_csv(r'C:\Users\rache\Downloads\us-marriages-divorces-1867-2014.csv')
marstat
```

```python
# get columns to plot
columns = marstat.columns.drop(['Marriages','Divorces','Population','Year'])

# create x data
x_data = marstat['Year']
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, marstat[column])
# set title and legend
ax.set_title('US Marriages and Divorces Per Capita')
ax.legend(columns)
ax.set_xlabel('Year')
ax.set_ylabel('Count')
```
